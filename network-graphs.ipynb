```python
def plot_network_graph(edges: pd.DataFrame,
                       source: str,
                       target: str,
                       directional=False,
                       edges_weight: str=None,
                       nodes: pd.DataFrame=None,
                       node_names: str=None,
                       nodes_weight: str=None,
                       nodes_category: str=None,
                       text_filter: str=None,
                       filter_degree_level: int=0,
                       width=900,
                       height=900,
                       seed=None
                       ):
    """
    Args:
         filter_degree_level:
            0: nodes of first-degree connections (but nodes are only connected to `text_filter`)
            1: in addition to `0`, add edges/connections between first-degree connections
            2: in addition to `1`, add edges and nodes to second-degree connections
    """
    # most of the code is adopted from https://github.com/rweng18/midsummer_network/blob/master/midsummer_graph.ipynb
    from sklearn.preprocessing import MinMaxScaler
    import plotly.graph_objects as go
    import networkx as nx


    assert nodes is None or node_names is not None

    edges = edges.copy()
    nodes = nodes.copy()

    def make_edge(edge_x, edge_y,
                  #text,
                  line_width, line_color):
        """
        Creates a scatter trace for the edge between x's and y's with given width

        Parameters
        ----------
        edge_x: a tuple of the endpoints' x-coordinates in the form, tuple([x0, x1, None])
        edge_y: a tuple of the endpoints' y-coordinates in the form, tuple([y0, y1, None])
        line_width: the width of the line
        line_color: color of the line

        Returns
        -------
        An edge trace that goes between x0 and x1 with specified width.
        """
        return  go.Scatter(
            x=edge_x,
            y=edge_y,
            line=dict(
                width=line_width,
                color=line_color
            ),
            hoverinfo='text',
            #text=([text]),
            mode='lines'
        )

    if text_filter:

        # get all edges where either the source or the target is equal to the text filter
        first_degree_edges = edges[(edges[source] == text_filter) | (edges[target] == text_filter)]
        # get a list of all the node names associated with edges
        first_degree_node_names = list(set(list(first_degree_edges[source]) + list(first_degree_edges[target])))

        if filter_degree_level == 0:
            # just add first-degree edges and nodes
            edges = first_degree_edges
            nodes = nodes[nodes[node_names].apply(lambda node_name: node_name in first_degree_node_names)]
        elif filter_degree_level == 1:
            # add connections/edges where the source and target are in the list of first_degree_node_names
            # meaning, edges where the source or the target equals the text_filter, or edges where the
            # source and target are in the first_degree_node_names.
            edges = edges[
                (edges[source] == text_filter) |
                (edges[target] == text_filter) |
                (
                    edges[source].apply(lambda value: value in first_degree_node_names) &
                    edges[target].apply(lambda value: value in first_degree_node_names)
                )
            ]
            nodes = nodes[nodes[node_names].apply(lambda node_name: node_name in first_degree_node_names)]
        elif filter_degree_level == 2:
            # here we are adding connections where the source and target are in the list of
            # first_degree_node_names (same as before), but now we are adding edges and nodes
            # where either the source or the target are in the first_degree_node_names (which gets
            # the second degree edges/nodes)
            edges = edges[
                # by definition this includes any edges where the source or target
                # is the text_filter
                # plus everyone else in the first-degree nodes
                edges[source].apply(lambda value: value in first_degree_node_names) |
                edges[target].apply(lambda value: value in first_degree_node_names)
            ]
            second_level_items = list(set(list(edges[source]) + list(edges[target])))
            nodes = nodes[nodes[node_names].apply(lambda node_name: node_name in second_level_items)]
        else:
            raise NotImplementedError()

    if edges_weight:
        edges['edges_size'] = MinMaxScaler(feature_range=(1, 5)).\
            fit_transform(pd.DataFrame(edges.loc[:, edges_weight]))
    else:
        edges['edges_size'] = 1

    if nodes_weight:
        nodes['nodes_size'] = MinMaxScaler(feature_range=(1, 5)).\
            fit_transform(pd.DataFrame(nodes.loc[:, nodes_weight]))
    else:
        nodes['nodes_size'] = 1

    node_color_lookup = None
    if nodes_category:
        color_values = [c.value for c in hlp.color.Colors]
        category_values = nodes[nodes_category].unique().tolist()
        category_values.sort()
        assert len(color_values) > len(category_values)
        node_color_lookup = {category: color for category, color in zip(category_values, color_values[0:len(category_values)])}
        del color_values, category_values

    graph = nx.Graph()
    # Add nodes to graph
    for row_index in range(0, nodes.shape[0]):
        graph.add_node(
            nodes.iloc[row_index][node_names],
            size = nodes.iloc[row_index]['nodes_size'],
            color=node_color_lookup[nodes.iloc[row_index]['category']] if nodes_category else 'gray',
            category=nodes.iloc[row_index]['category'],
        )

    # For each co-appearance between two nodes, add an edge
    for row_index in range(0, edges.shape[0]):
        instance = edges.iloc[row_index]
        graph.add_edge(instance[source],
                       instance[target],
                       weight = instance['edges_size'])

    positions = nx.spring_layout(graph, seed=seed)

    edge_trace = []
    for edge in graph.edges():
        if graph.edges()[edge]['weight'] > 0:
            source_value = edge[0]
            target_value = edge[1]

            x0, y0 = positions[source_value]
            x1, y1 = positions[target_value]
            #text = source_value + '--' + target_value + ': ' + str(graph.edges()[edge]['weight'])

            if text_filter and directional:
                colors = 'red' if source_value == text_filter else 'blue'
            else:
                colors = 'blue'

            trace  = make_edge(
                edge_x=[x0, x1, None],
                edge_y=[y0, y1, None],
                #text=text,
                line_width=0.3 * graph.edges()[edge]['weight'] ** 1.75,
                line_color=colors
            )
            edge_trace.append(trace)

    # Make a node trace
    node_trace = go.Scatter(
        x=[],
        y=[],
        text=[],
        textposition="top center",
        textfont={'size': 10},
        mode='markers+text',
        hoverinfo='none',
        marker=dict(color=[],
                    size=[],
                    line=None)
    )
    # For each node in midsummer, get the position and size and add to the node_trace
    for node in graph.nodes():
        x, y = positions[node]
        node_trace['x'] += tuple([x])
        node_trace['y'] += tuple([y])
        #    print(graph.nodes()[node]['color'])
        node_trace['marker']['color'] += tuple([graph.nodes()[node]['color']])
        node_trace['marker']['size'] += tuple([5*graph.nodes()[node]['size']])
        node_text = '<b>' + node + '</b>'
        if nodes_category:
            node_text += '</br></br><sup>' + graph.nodes()[node]['category'] + '</sup>'
        node_trace['text'] += tuple([node_text])

    layout = go.Layout(
        paper_bgcolor='rgba(0,0,0,0)',
        plot_bgcolor='rgba(0,0,0,0)'
    )
    fig = go.Figure(layout = layout)

    for trace in edge_trace:
        fig.add_trace(trace)

    fig.add_trace(node_trace)
    fig.update_layout(
        showlegend = False,
        width=width,
        height=height
    )
    fig.update_xaxes(showticklabels = False)
    fig.update_yaxes(showticklabels = False)
    return fig

plot_network_graph(
    edges=edges_df,
    source='source',
    target='target',
    directional=False,
    edges_weight='weight',
    nodes=node_df,
    node_names='name',
    nodes_weight='weight',
    nodes_category='category',
    text_filter='Petyr-Baelish',
    filter_degree_level=0,
    width=1000,
    height=1000,
    seed=42,
)
```

```python
def plot_network_graph(edges: pd.DataFrame,
                       source: str,
                       target: str,
                       # directional=False,
                       edges_weight: str=None,
                       nodes: pd.DataFrame=None,
                       node_names: str=None,
                       nodes_weight: str=None,
                       nodes_category: str=None,
                       # width=900,
                       # height=900,
                       seed=None
                       ):
    import math

    def add_edge(edge_start, edge_end, edge_x_list, edge_y_list, length_frac=1,
                 arrow_pos = None, arrow_length = 0.025, arrow_angle = 30, dot_size = 20):
        # edge_start and edge_end are lists defining start and end points
        # Edge x and y are lists used to construct the graph
        # arrowAngle and arrowLength define properties of the arrowhead
        # arrowPos is None, 'middle' or 'end' based on where on the edge you want the arrow to appear
        # arrowLength is the length of the arrowhead
        # arrowAngle is the angle in degrees that the arrowhead makes with the edge
        # dotSize is the plotly scatter dot size you are using (used to even out line spacing when you have a mix of edge lengths)

        # Get start and end cartesian coordinates
        x0, y0 = edge_start
        x1, y1 = edge_end

        # Incorporate the fraction of this segment covered by a dot into total reduction
        length = math.sqrt( (x1-x0)**2 + (y1-y0)**2 )
        dot_size_conversion = .0565/20 # length units per dot size
        converted_dot_diameter = dot_size * dot_size_conversion
        length_frac_reduction = converted_dot_diameter / length
        length_frac = length_frac - length_frac_reduction

        # If the line segment should not cover the entire distance, get actual start and end coords
        skip_x = (x1-x0)*(1-length_frac)
        skip_y = (y1-y0)*(1-length_frac)
        x0 = x0 + skip_x/2
        x1 = x1 - skip_x/2
        y0 = y0 + skip_y/2
        y1 = y1 - skip_y/2

        # Append line corresponding to the edge
        edge_x_list.append(x0)
        edge_x_list.append(x1)
        edge_x_list.append(None) # Prevents a line being drawn from end of this edge to start of next edge
        edge_y_list.append(y0)
        edge_y_list.append(y1)
        edge_y_list.append(None)

        # Draw arrow
        if arrow_pos:
            # Find the point of the arrow; assume is at end unless told middle
            pointx = x1
            pointy = y1

            eta = math.degrees(math.atan((x1-x0)/(y1-y0))) if y1!=y0 else 90.0

            if arrow_pos == 'middle' or arrow_pos == 'mid':
                pointx = x0 + (x1-x0)/2
                pointy = y0 + (y1-y0)/2

            # Find the directions the arrows are pointing
            signx = (x1-x0)/abs(x1-x0) if x1!=x0 else +1    #verify this once
            signy = (y1-y0)/abs(y1-y0) if y1!=y0 else +1    #verified

            # Append first arrowhead
            dx = arrow_length * math.sin(math.radians(eta + arrow_angle))
            dy = arrow_length * math.cos(math.radians(eta + arrow_angle))
            edge_x_list.append(pointx)
            edge_x_list.append(pointx - signx ** 2 * signy * dx)
            edge_x_list.append(None)
            edge_y_list.append(pointy)
            edge_y_list.append(pointy - signx ** 2 * signy * dy)
            edge_y_list.append(None)

            # And second arrowhead
            dx = arrow_length * math.sin(math.radians(eta - arrow_angle))
            dy = arrow_length * math.cos(math.radians(eta - arrow_angle))
            edge_x_list.append(pointx)
            edge_x_list.append(pointx - signx ** 2 * signy * dx)
            edge_x_list.append(None)
            edge_y_list.append(pointy)
            edge_y_list.append(pointy - signx ** 2 * signy * dy)
            edge_y_list.append(None)


        return edge_x_list, edge_y_list

    from sklearn.preprocessing import MinMaxScaler
    import plotly.graph_objects as go
    import networkx as nx

    if edges_weight:
        edges['edges_size'] = MinMaxScaler(feature_range=(1, 3)).\
            fit_transform(pd.DataFrame(edges.loc[:, edges_weight]))
    else:
        edges['edges_size'] = 1

    if nodes_weight:
        nodes['nodes_size'] = MinMaxScaler(feature_range=(1, 5)).\
            fit_transform(pd.DataFrame(nodes.loc[:, nodes_weight]))
    else:
        nodes['nodes_size'] = 1

    node_color_lookup = None
    if nodes_category:
        color_values = [c.value for c in hlp.color.Colors]
        category_values = nodes[nodes_category].unique().tolist()
        category_values.sort()
        assert len(color_values) > len(category_values)
        node_color_lookup = {category: color for category, color in zip(category_values, color_values[0:len(category_values)])}
        del color_values, category_values

    graph = nx.Graph()
    # Add nodes to graph
    for row_index in range(0, nodes.shape[0]):
        graph.add_node(
            nodes.iloc[row_index][node_names],
            size = nodes.iloc[row_index]['nodes_size'],
            color=node_color_lookup[nodes.iloc[row_index]['category']] if nodes_category else 'gray',
            category=nodes.iloc[row_index]['category'],
        )

    # For each co-appearance between two nodes, add an edge
    for row_index in range(0, edges.shape[0]):
        instance = edges.iloc[row_index]
        graph.add_edge(instance[source],
                       instance[target],
                       weight = instance['edges_size'])

    positions = nx.spring_layout(graph, seed=seed)

    # Controls for how the graph is drawn
    node_color = 'black'
    line_width = 1

    node_x = []
    node_y = []
    node_size = [x*10 for x in nodes.nodes_size]

    for node in graph.nodes():
        x, y = positions[node]
        node_x.append(x)
        node_y.append(y)

    # Make a list of edges for plotly, including line segments that result in arrowheads
    #edge_color = ['red' if x == word_filter else 'blue' for x in filtered_network_data['source']]
    edge_traces = []
    for edge in graph.edges():
        edge_x = []
        edge_y = []
        # addEdge(start, end, edge_x, edge_y, lengthFrac=1, arrowPos = None, arrowLength=0.025, arrowAngle = 30, dotSize=20)
        char_1 = edge[0]
        char_2 = edge[1]
        start = list(positions[char_1])
        end = list(positions[char_2])
        edge_x, edge_y = add_edge(start, end, edge_x, edge_y, 1, 'end', .02, 30, 1)


        word_filter = 'update'
        edge_traces += [go.Scatter(x=edge_x, y=edge_y,
                                   opacity=0.3,
                                   line=dict(width=line_width, color='red' if char_1 == word_filter else 'blue'),
                                   hoverinfo='none', mode='lines')]
    node_trace = go.Scatter(
        x=node_x,
        y=node_y,
        mode='markers',
        #text=nodes['app'].values,
        #textposition='top center',
        #textfont=dict(size=10, color='black'),
        opacity=0.5,
        hoverinfo='text',
        marker=dict(showscale=False, color = node_color, size=node_size)#size=nodeSize)
    )
    text_trace = go.Scatter(
        x=node_x,
        y=node_y,
        mode='text',
        text=[f'{x}<br><sup>{y}</sup>' for x, y in zip(nodes[node_names], nodes[nodes_category])],
        textposition='top center',
        textfont=dict(size=11, color='black'),
        #opacity=0.6,
        #    hoverinfo='text',
        #    marker=dict(showscale=False, color = nodeColor, size=nodes_size)#size=nodeSize)
    )

    fig = go.Figure(data=edge_traces + [node_trace, text_trace],
                    layout=go.Layout(
                        showlegend=False,
                        hovermode='closest',
                        margin=dict(b=20,l=5,r=5,t=40),
                        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                    )

    # Note: if you don't use fixed ratio axes, the arrows won't be symmetrical
    fig.update_layout(yaxis = dict(scaleanchor = "x", scaleratio = 1),
                      plot_bgcolor='rgb(255,255,255)',
                      width=900,
                      height=900)

    return fig
```
