```python
import requests
from bs4 import BeautifulSoup

info = "https://omscs.gatech.edu/current-courses"
response = requests.get(info)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    section = soup.find('h3', text="Current & Ongoing OMS Courses").find_next('ul')
    link_elements = section.find_all('a')
    courses = {link.text.strip(): {'url': link.get('href')} for link in link_elements if link.get('href')}
    for name, info in courses.items():
        print(f"{name}: {info['url']}")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
# # Scrape the "Overview" and "Before Taking This Class... Suggested Background Knowledge" sections
# for name, info in courses.items():
#     print(name)
#     response = requests.get(info['url'])
#     if response.status_code == 200:
#         course_html = response.text
#         course_soup = BeautifulSoup(course_html, 'html.parser')
#         # Find the "Overview" section
#         overview_section = course_soup.find('h4', string="Overview")
#         if overview_section:
#             overview_text = overview_section.find_next('p').text
#             courses[name]['overview'] = overview_text
#             print(f"Overview:\n{overview_text}")

#         # Find the "Before Taking This Class... Suggested Background Knowledge" section
#         suggested_background_section = course_soup.find('h4', string="Before Taking This Class...")#.find_next('h6')
#         if suggested_background_section:
#             suggested_background_text = suggested_background_section.find_next('p').text
#             courses[name]['suggested_background'] = suggested_background_text
#             print(f"Suggested Background Knowledge:\n{suggested_background_text}")
#     else:
#         print(f"Failed to retrieve course page for {name}. Status code: {response.status_code}")
#     print("\n----\n")
# Scrape the "Overview" and "Before Taking This Class... Suggested Background Knowledge" sections
for name, info in courses.items():
    print(name)
    response = requests.get(info['url'])
    if response.status_code == 200:
        course_html = response.text
        course_soup = BeautifulSoup(course_html, 'html.parser')
        # Find the "Overview" section
        overview_section = course_soup.find('h4', string="Overview")
        if overview_section:
            overview_text = ""
            current_element = overview_section.find_next_sibling()
            while current_element and current_element.name != 'h4':
                if current_element.name == 'p':
                    overview_text += current_element.text + "\n"
                elif current_element.name == 'ul':
                    for item in current_element.find_all('li'):
                        overview_text += item.text + "\n"
                current_element = current_element.find_next_sibling()
            courses[name]['overview'] = overview_text.strip()
            print(f"Overview:\n{overview_text.strip()}")

        # Find the "Before Taking This Class... Suggested Background Knowledge" section
        suggested_background_section = course_soup.find('h4', string="Before Taking This Class...")
        if suggested_background_section:
            suggested_background_text = suggested_background_section.find_next('p').text
            courses[name]['suggested_background'] = suggested_background_text
            print(f"Suggested Background Knowledge:\n{suggested_background_text}")
    else:
        print(f"Failed to retrieve course page for {name}. Status code: {response.status_code}")
    print("\n----\n")
```

```python
# for url in specializations:
#     response = requests.get(url)
#     if response.status_code == 200:
#         soup = BeautifulSoup(response.text, 'html.parser')
#         core_courses_heading = soup.find('h3', text=lambda x: 'Core Courses' in x)
        
#         if core_courses_heading:
#             core_courses_text = ""
#             for sibling in core_courses_heading.find_next_siblings():
#                 if sibling.name == 'h3':
#                     break
#                 core_courses_text += sibling.get_text(strip=True) + "\n"
            
#             print("URL:", url)
#             print("Core Courses Text:")
#             print(core_courses_text)
#             print("\n" + "="*50 + "\n")
#         else:
#             print("Core Courses not found on:", url)
#     else:
#         print("Failed to retrieve the page:", url)

for name, info in specializations.items():
    response = requests.get(info['url'])
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        core_courses_heading = soup.find('h3', text=lambda x: 'Core Courses' in x)
        
        if core_courses_heading:
            core_courses_text = ""
            next_sibling = core_courses_heading.find_next_sibling()
            
            while next_sibling and next_sibling.name != 'h3':
                if next_sibling.name == 'ul':
                    list_items = next_sibling.find_all('li')
                    for item in list_items:
                        core_courses_text += item.get_text() + '\n'
                elif next_sibling.name == 'p':
                    core_courses_text += next_sibling.get_text() + '\n'
                else:
                    core_courses_text += next_sibling.string + '\n'
                next_sibling = next_sibling.find_next_sibling()
            
            print("URL:", info['url'])
            print("Core Courses Markdown:")
            specializations[name]['core_courses'] = core_courses_text
            print(core_courses_text)
            print("\n" + "="*50 + "\n")
        else:
            print("Core Courses not found on:", info['url'])
    else:
        print("Failed to retrieve the page:", info['url'])
```

```python
for name, info in specializations.items():
    response = requests.get(info['url'])
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        electives_heading = soup.find('h3', text=lambda x: 'Electives' in x)
        
        if electives_heading:
            electives_text = ""
            next_sibling = electives_heading.find_next_sibling()
            
            while next_sibling and next_sibling.name != 'h3':
                if next_sibling.name == 'p':
                    electives_text += next_sibling.get_text() + '\n'
                elif next_sibling.name == 'ul':
                    list_items = next_sibling.find_all('li')
                    for item in list_items:
                        if item.find('a'):
                            course_link = item.find('a')['href']
                            course_name = item.find('strong').get_text(strip=True)
                            electives_text += f"{course_name} [{course_link}]"
                        else:
                            electives_text += item.get_text(strip=True)
                        electives_text += '\n'
                else:
                    electives_text += next_sibling.string + '\n'
                next_sibling = next_sibling.find_next_sibling()
            
            print("URL:", info['url'])
            print("Electives Markdown:")
            print(electives_text)
            specializations[name]['electives'] = electives_text
            print("\n" + "="*50 + "\n")
        else:
            print("Electives not found on:", info['url'])
    else:
        print("Failed to retrieve the page:", info['url'])
```
